Problem 1: The Task Manager (List Practice)
Your goal is to manage a simple list of tasks.

Problem 5: The Gradebook (Combining)
You have a list of student names and you want to create a default gradebook.

Start with this list of names: students = ['Alice', 'Bob', 'Charlie']

Create an empty dictionary called gradebook.

Loop through the students list. For each student, add them to the gradebook as a key. Set their value to 0 (a default grade).

Print the final gradebook dictionary.

```python
students = ['Alice', 'Bob', 'Charlie']

gradebook={}
for st in students:
    gradebook[st]=0
print(gradebook)    
```

Problem 1: The E-Commerce Data Cleaner (Lists & Dictionaries)
You have a list of dictionaries representing products. The data is messy. One product is missing the "in_stock" key. Your task is to get the names of all products that are in stock.

Rule: A product is considered "out of stock" if in_stock is False OR if the in_stock key is missing entirely.

Your Data:

Create: Make a list named shopping_list containing: "apples", "bananas", "milk", "bread".

Read (Index): Print the second item in the list (remember, indexing starts at 0!).

Read (Slice): Create a new list essentials that contains only "milk" and "bread" by slicing the shopping_list. Print essentials.

Update: Your store is out of "bananas". Change the "bananas" item to "blueberries".

Add (Append): You forgot something. Add "eggs" to the end of shopping_list using .append().

Add (Insert): You want "coffee" to be the first item. Use .insert() to add it at the correct position.

Delete (Remove): You've decided against "milk". Remove it from the list by its value using .remove().

Delete (Pop): Remove the last item from the list using .pop() and print the item that was removed.

Check: Print your final shopping_list.

```python
shopping_list = ["apples", "bananas", "milk", "bread"]

# 1. Create (Already done)
# 2. Read (Index)
print(shopping_list[1])  # Prints: bananas

# 3. Read (Slice)
essentials = shopping_list[2:4]
print(essentials)  # Prints: ['milk', 'bread']

# 4. Update
shopping_list[1] = "blueberries"
# List is now: ['apples', 'blueberries', 'milk', 'bread']

# 5. Add (Append)
shopping_list.append("eggs")
# List is now: ['apples', 'blueberries', 'milk', 'bread', 'eggs']

# 6. Add (Insert) - CORRECTED
shopping_list.insert(0, "coffee")
# List is now: ['coffee', 'apples', 'blueberries', 'milk', 'bread', 'eggs']

# 7. Delete (Remove) - CORRECTED
shopping_list.remove("milk")
# List is now: ['coffee', 'apples', 'blueberries', 'bread', 'eggs']

# 8. Delete (Pop)
removed_item = shopping_list.pop()
print(f"Removed item: {removed_item}") # Prints: Removed item: eggs

# 9. Check
print(shopping_list)
```

Create (from list): Create a list duplicate_nums = [1, 2, 3, 2, 4, 5, 1, 3].

Remove Duplicates: Create a new set unique_nums from duplicate_nums. Print unique_nums to see the duplicates vanish.

Add: Add the number 6 to unique_nums using .add().

Check Membership: Write one line of code that prints True if 4 is in unique_nums, and False otherwise.

Set Operations:

Create another set: more_nums = {4, 5, 6, 7, 8}.

Find the intersection (items in both sets) and print it.

```python
duplicate_nums=[1,2,3,2,4,5,1,3]
set_unique=set(duplicate_nums)
print(set_unique)
set_unique.add(6)
print(4 in set_unique)
more_nums={4,5,6,7,8}

# set_unique = {1, 2, 3, 4, 5, 6}
# more_nums  = {4, 5, 6, 7, 8}

# 1. Intersection (items in BOTH)
print("Intersection:", set_unique.intersection(more_nums))

# 2. Union (items in EITHER)
print("Union:", set_unique.union(more_nums))

# 3. Difference (items in set_unique but NOT in more_nums)
print("Difference:", set_unique.difference(more_nums))
# Or, using the - operator: print(set_unique - more_nums)
```
