Problem 1: The Task Manager (List Practice)
Your goal is to manage a simple list of tasks.

Create an empty list called tasks.

Add the following three tasks to the list (as strings): "Pay bills", "Do laundry", "Call mom".

You just remembered a high-priority task! Insert "Go to gym" at the beginning of the list.

You just finished the last task on your list. Remove it and store it in a variable called completed_task.

Print the completed_task.

Print the final tasks list.

```python
tasks=[]
tasks.append("Pay bills")
tasks.append("Do laundry")
tasks.append("Call mom")
tasks.insert(0,"Go to gym")

completed_task=tasks.pop()
print(f"Completed Task: {completed_task}")
print(f"Updated TAsk: {tasks}")
```

Problem 2: The Guest List (Set Practice)
You are managing RSVPs for a party. You have a list of people who have replied, but some have replied more than once.

Start with this list: rsvps = ['Alice', 'Bob', 'Charlie', 'David', 'Alice', 'Eve', 'Bob']

How many total RSVPs did you receive? (Print the number).

How many unique people are coming to the party? (Print the number). Hint: Use a set to find this.

```python
rsvps = ['Alice', 'Bob', 'Charlie', 'David', 'Alice', 'Eve', 'Bob']

unique_rsvp=set(rsvps)
print(f"Total RSVPs: {len(rsvps)}")
print(f"Unique rsvp: {len(unique_rsvp)}")
```

Problem 3: The Inventory Check (Dictionary Practice)
You are managing a single product in your e-commerce store.

Create a dictionary called product with the following information:

"name": "Laptop"

"price": 1200

"in_stock": True

Print the product's price by accessing it with its key.

The price just changed! Update the price to 1350.

You just sold out. Update the in_stock value to False.

Print the final product dictionary.

```python
product={"name":"Laptop",
         "price":1200,
         "in_stock":True
         }
print(product["price"])
product["price"]=1350
product["in_stock"]=False
print(product)
```

Problem 4: The Map Coordinates (Tuple Practice)
You are storing fixed map coordinates. A user's "Home" and "Work" locations will never change.

Create a tuple called home_coords with the values (40.7128, -74.0060).

Create a tuple called work_coords with the values (40.7580, -73.9855).

Create a dictionary called locations. Use the tuples home_coords and work_coords as the keys.

Set the value for home_coords to be the string "Home".

Set the value for work_coords to be the string "Work".

Print the locations dictionary. This tests the "killer feature" of tuples: being used as dictionary keys.

```python
home_coords=(40.7128,-74.0060)
work_coords=(40.7580,-73.9855)
locations={
    home_coords:"Home",
    work_coords:"Work"

}

print(f"Location Dict : {locations}")
```

Problem 5: The Gradebook (Combining)
You have a list of student names and you want to create a default gradebook.

Start with this list of names: students = ['Alice', 'Bob', 'Charlie']

Create an empty dictionary called gradebook.

Loop through the students list. For each student, add them to the gradebook as a key. Set their value to 0 (a default grade).

Print the final gradebook dictionary.

```python
students = ['Alice', 'Bob', 'Charlie']

gradebook={}
for st in students:
    gradebook[st]=0
print(gradebook)    
```
