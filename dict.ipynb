A dictionary (or dict) is a data type that stores information in keyâ€“value pairs.

Think of it like a real-world dictionary:

A word = the key

Its definition = the value

```python
person={
    "name":"Aditya",
    "age":22,
    "city":"HYD"
}

print(person)
```

```python
print(person["name"])
print(person["age"])

person["height"]=170

print(person["height"])

print(person)

del person["height"]

print(person)
```

```python
for key,value in person.items():
    print(key,":",value)
```
