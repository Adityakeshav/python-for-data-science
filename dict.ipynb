A dictionary (or dict) is a data type that stores information in key–value pairs.

Think of it like a real-world dictionary:

A word = the key

Its definition = the value

```python
person={
    "name":"Aditya",
    "age":22,
    "city":"HYD"
}

print(person)
```

```python
print(person["name"])
print(person["age"])

person["height"]=170

print(person["height"])

print(person)

del person["height"]

print(person)
```

```python
for key,value in person.items():
    print(key,":",value)
```

In Python, a defaultdict is a special kind of dictionary that automatically gives a default value to any key that doesn’t exist yet.

```python
from collections import defaultdict

count=defaultdict(int)

count["apple"]+=1
count["apple"]+=1
count["banana"]+=1

print(count)
```
